```python
import seaborn as sns

outliers_cols = [
    'cons_12m',
    'cons_gas_12m',
    'cons_last_month',
    'forecast_cons_12m',
    'forecast_cons_year',
    'forecast_discount_energy',
    'forecast_meter_rent_12m',
    'forecast_price_energy_off_peak',
    'imp_cons',
    'margin_gross_pow_ele',
    'margin_net_pow_ele',
    'pow_max'
]

for i in outliers_cols:
  plt.figure(figsize=(4,4))
  sns.histplot(df[i] , kde = True)
  plt.title(i)
  plt.show()
```

```python
cat_col = ['channel_sales' , 'has_gas' , 'origin_up']

for i in cat_col:
  print('-' * 10)
  print(df[i].value_counts(normalize=True)*100)
```

• channel_sales contains eight categories, including a "MISSING" category representing approximately 25.5% of the records.

• has_gas is highly imbalanced, with nearly 82% of customers not having gas service.

• origin_up contains six acquisition sources plus a small "MISSING" category. More than 99% of customers belong to the top three acquisition channels.

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(numeric_only=True) , annot=True , fmt = '.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.corr(numeric_only=True)['churn'].sort_values(key=abs, ascending=False)
```

Observations

• None of the numerical features show a strong linear correlation with churn.

• margin_net_pow_ele and margin_gross_pow_ele have the highest positive correlation with churn (≈0.10), but the relationship is weak.

• num_years_antig shows a weak negative correlation (-0.07), suggesting customers with longer tenure may be slightly less likely to churn.

• Most numerical variables have correlation values close to zero, indicating that customer churn is likely influenced by multiple interacting factors rather than a single feature.
