```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:,.2f}'.format
```

```python
clean = pd.read_csv('clean_data_after_eda.csv')
clean.head(2)
```

```python
clean.columns
```

```python
clean.describe().T
```

```python
L = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal']

for i in L:
 clean[i] = pd.to_datetime(clean[i])
```

```python
clean.info()
```

```python
clean['avg_monthly_cons'] = clean['cons_12m']/12
```

monthly scale of yearly consumption

```python
clean['consumption_deviation'] = clean['cons_last_month'] = clean['avg_monthly_cons']
```

checking if last month consumption is unusual or not

```python
clean['forecast_error'] = clean['forecast_cons_12m'] - clean['cons_12m']
```

gap between forcasted and actual

```python
clean['contract_duration'] = clean['date_end'] - clean['date_activ']
```

checking for the long contract customer

```python
clean['days_to_modify_product'] = (
    clean['date_modif_prod'] - clean['date_activ']
).dt.days
```

how fast customer modify product or service

```python
clean['margin_difference'] = (
    clean['margin_gross_pow_ele']
    - clean['margin_net_pow_ele']
)
```

gap between gross and net margin

```python
clean['consumption_per_power'] = (
    clean['cons_12m'] / clean['pow_max'].replace(0 , np.nan)
)
```

usage according to subscribed power

```python
price = pd.read_csv('price_data (1).csv')
price.head(2)
```

```python
price['price_date'] = pd.to_datetime(price['price_date'])
```

```python
price.head()
```

```python
price.info()
```

```python
price.columns
```

```python
price['month'] = price['price_date'].dt.month
```

```python
price[['price_date','month']].head()
```

```python
jan = price[price['month'] == 1]
dec = price[price['month'] == 12]
```

```python
jan = jan[['id','price_off_peak_var']]
dec = dec[['id','price_off_peak_var']]
```

```python
jan.rename(columns={'price_off_peak_var':'jan_price'}, inplace=True)
dec.rename(columns={'price_off_peak_var':'dec_price'}, inplace=True)
```

```python
price_diff = jan.merge(dec,on='id')
```

```python
price_diff['off_peak_diff_dec_jan'] = (
    price_diff['dec_price'] -
    price_diff['jan_price']
)
```

```python
clean = clean.merge(
    price_diff[['id','off_peak_diff_dec_jan']],
    on='id',
    how='left'
)
```
